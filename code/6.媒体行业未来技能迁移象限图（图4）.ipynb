```python
plt.figure(figsize=(16,10), dpi= 80)

x,y= df_four['未来重要性水平'], df_four['现在重要性水平']
label = df_four['技能名称']
plt.scatter(x, y,s=70.,c='b')
plt.xlabel('未来重要性水平',fontsize=22)
plt.ylabel('现在重要性水平',fontsize=22)

for a,b,l in zip(x,y,label): # zip 拉链函数将其配对组合
    plt.text(a, b, '%s.' % l, ha='left', va='bottom', 
                                             fontsize=19)
plt.xlim(-0.1,1.1)    
plt.ylim(-0.1,1.1)  
plt.tick_params(labelsize=16)
plt.vlines(x=0.5, ymin=-0.1, ymax=1.1, 
           colors='red', linewidth=1)
plt.hlines(y=0.5, xmin=-0.1, xmax=1.1,
           colors='red', linewidth=1)
```
