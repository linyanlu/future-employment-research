```python
def R2(y_test, y_pred): 
    residuals_sum_of_squares = np.sum((y_pred - y_test)**2)
    total_sum_of_squares = np.sum((y_test - np.mean(y_test))**2)
    return 1 - residuals_sum_of_squares/total_sum_of_squares
```
