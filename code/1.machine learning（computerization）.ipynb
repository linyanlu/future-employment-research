```python
import pandas as pd
import numpy as np
```

```python
df_1=open('/Users/Asus/Desktop/文件夹/项目/人工智能时代媒体就业/data/onet_db_23_2_text/Skills.txt')
df_s = pd.read_csv(df_1, sep = '\t')
df_s=df_s[df_s['Scale ID']=='LV']
df_s=df_s[df_s['Element Name'].isin(['Social Perceptiveness','Negotiation','Persuasion'])]
# df_s=df_s[['O*NET-SOC Code','Element Name','Data Value']]
# df_s.head()

## len(df_s['O*NET-SOC Code'])
df_2=pd.read_excel('/Users/Asus/Desktop/文件夹/项目/人工智能时代媒体就业/data/onet_db_23_2_text/2010_to_2018_SOC_Crosswalk.xlsx')
##df_2.head()
df_2=df_2.rename(columns={'O*NET-SOC 2010 Code':'O*NET-SOC Code'})
df_s1=df_s.merge(df_2,how='left',on='O*NET-SOC Code')
##df_s1
df_s1=df_s1.rename(columns={'2018 SOC Code':'soc'})
df_3=pd.read_excel('/Users/Asus/Desktop/文件夹/项目/人工智能时代媒体就业/data/onet_db_23_2_text/future_of_employment_appendix_data.xlsx')
df_3.head()
df_s2=df_3.merge(df_s1,how='left',on='soc')
df_s2.head()
```

```python

df_a=open('/Users/Asus/Desktop/文件夹/项目/人工智能时代媒体就业/data/onet_db_23_2_text/Abilities.txt')
df_a = pd.read_csv(df_a, sep = '\t')
df_a=df_a[df_a['Scale ID']=='LV']
df_a=df_a[df_a['Element Name'].isin(['Finger Dexterity','Manual Dexterity','Originality'])]
# df_a=df_a[['O*NET-SOC Code','Element Name','Data Value']]
# df_a.head()
df_a1=df_a.merge(df_2,how='left',on='O*NET-SOC Code')
df_a1=df_a1.rename(columns={'2018 SOC Code':'soc'})
df_a2=df_3.merge(df_a1,how='left',on='soc')
df_a2.head()
```

```python

df_k=open('/Users/Asus/Desktop/文件夹/项目/人工智能时代媒体就业/data/onet_db_23_2_text/Knowledge.txt')
df_k = pd.read_csv(df_k, sep = '\t')
df_k=df_k[df_k['Scale ID']=='LV']
df_k=df_k[df_k['Element Name']=='Fine Arts']

# df_a.head()
df_k1=df_k.merge(df_2,how='left',on='O*NET-SOC Code')
df_k1=df_k1.rename(columns={'2018 SOC Code':'soc'})
df_k2=df_3.merge(df_k1,how='left',on='soc')
df_k2.head()
```

```python
df_w=open('/Users/Asus/Desktop/文件夹/项目/人工智能时代媒体就业/data/onet_db_23_2_text/Work Activities.txt')
df_w = pd.read_csv(df_w, sep = '\t')
df_w=df_w[df_w['Scale ID']=='IM']
df_w=df_w[df_w['Element Name']=='Assisting and Caring for Others']

# df_a.head()
df_w1=df_w.merge(df_2,how='left',on='O*NET-SOC Code')
df_w1=df_w1.rename(columns={'2018 SOC Code':'soc'})
df_w2=df_3.merge(df_w1,how='left',on='soc')
df_w2.head()
```

```python
df_c=open('/Users/Asus/Desktop/文件夹/项目/人工智能时代媒体就业/data/onet_db_23_2_text/Work Context.txt')
df_c = pd.read_csv(df_c, sep = '\t')
df_c=df_c[df_c['Element Name']=='Cramped Work Space, Awkward Positions']
df_c=df_c[df_c['Scale ID']=='CX']
df_c
df_c1=df_c.merge(df_2,how='left',on='O*NET-SOC Code')
df_c1=df_c1.rename(columns={'2018 SOC Code':'soc'})
df_c2=df_3.merge(df_c1,how='left',on='soc')
df_c2.head()
```

```python
df = pd.concat( [df_s2, df_a2,df_k2,df_w2,df_c2], axis=0 )
df=df.sort_values(by=['O*NET-SOC Code','Element Name'])

df
df_total=df[['id','label','soc','O*NET-SOC Code','Element Name','Data Value','computerization']]
df_total
```

```python
import numpy as np
df_70=df_total.dropna(subset=["label"])
df_70
len(df_70['soc'].unique())
df_70_1=df_70.groupby(['soc','Element Name']).mean()
df_70_1.reset_index(inplace=True)
# df_70_1.to_csv("data.csv")
# df_70_1
# len(df_70_1['soc'].unique())
```

```python
import pandas as pd 
import pylab as plt
import seaborn as sns
import GPy
import numpy as np
```

```python
data_list=list(df_70_1['Data Value'])
X=[]
for i in range(0,585,9):
    list1=data_list[i:i+9]
    X.append(list1)
len(X)
```

```python
X=np.array(X)
# X
```

```python
data_list1=list(df_70_1['label'])
Y=[]
for i in range(0,585,9):
    list1=data_list1[i]
    Y.append(list1)
Y=np.array(Y)
Y=Y[:,np.newaxis]
# Y
```

```python
kernel = GPy.kern.RBF(input_dim=9, variance=1., lengthscale=1.)
```

```python
m = GPy.models.GPRegression(X,Y,kernel)
```

```python
m.optimize(messages=True)
```

```python
print(m)
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score 
X1, X2, y1, y2 = train_test_split(X, Y, random_state=0,
                                  train_size=0.6, test_size = 0.4)
m = GPy.models.GPRegression(X1,y1,kernel)
m.optimize(messages=True)
y2_model = m.predict(X2)[0]
len(y2_model)

for i in range(len(y2_model)):
        if y2_model[i]>0.5:
            y2_model[i]=1
        else:
            y2_model[i]=0
    
accuracy_score(y2, y2_model)
```
